```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Lecture du dataset nettoyé
df = pd.read_csv("/Users/germain.pierroz/Documents/apprentissage_python/mon_projet_ml/data/dataset_clean.csv")  # Assure-toi que le chemin est correct
df.head()
```

```python
# Affichage des statistiques descriptives
df.describe()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], kde=True, bins=30)
plt.title("Distribution de l'Age")
plt.xlabel("Age (standardisé)")
plt.ylabel("Fréquence")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Fare'], kde=True, bins=30)
plt.title("Distribution de Fare")
plt.xlabel("Fare (standardisé)")
plt.ylabel("Fréquence")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette='coolwarm')
plt.title("Relation entre Age et Fare")
plt.xlabel("Age (standardisé)")
plt.ylabel("Fare (standardisé)")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Sex', y='Age')
plt.title("Distribution de l'Age selon le Sexe\n(0: female, 1: male)")
plt.xlabel("Sex")
plt.ylabel("Age (standardisé)")
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'Sex', 'Survived']])
plt.show()
```
